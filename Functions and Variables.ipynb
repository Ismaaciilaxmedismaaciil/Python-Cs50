### Integer (int), and Operations

```python
# Simple Mathmetical Operation

# Addition (+)
# Substaraction (-)
# Multiplication (*)
# Division (/)
# Modulus (%)
# Floor Division (//)
# Exponension or Square (**)
```

```python
#  Simple Calculator
num1 = float(input('Enter first number: '))
num2 = float(input('Enter second number: '))

# Addition the two numbers
add = num1 + num2

# Substraction
subs = num1 - num2

# Multiplication
mult = num1 * num2

# Division
div = num1 / num2

# Modulus
remainder = num1 % num2

# Floor division
floor_div = num1 // num2

# Exponension or Square
exp = num1 ** num2

# Printing results
print(f'Addtion: {add}')
print(f'Substraction: {subs}')
print(f'Multiplication: {mult}')
print(f'Division: {div}')
print(f'Remainder / Modulus: {remainder}')
print(f'Floor division: {floor_div}')
print(f'Exponension / Square: {exp}')
```

```python
# Addition Operator in integer type
x = int(3.1)
y = int(3)

print(x + y)
```

```python
# Addition Operator in float type
x = float(3.1)
y = float(3)

print(x + y)
```

### Rounding Numbers
 - RoundUp
 - RoundDown

```python
# Round-Up and Down
# Example

x = float(input("What's X? "))
y = float(input("What's Y? "))

#Roundin the result by adding, and than putting z variable
z = round(x + y)

print(z)
```

```python
# Using ',' for every trible digits. like(1,000)
x = 999
y = 1

z = round(x + y)

print(f'{z:,}')
```

```python
# Example 2: 
x = 9999
y = 100

z = round(x + y)

print(f'{z:,}')
```

```python
# Division Operator 
x = 100
y = 5

z = x / y

print(z)
```

```python
# Rounding the second digit
x = 2
y = 3

z = round(x / y, 2) # round the second digit

print(z)
```

```python
# Division Operator 
x = 2
y = 3

z = x / y

print(f'{z:.2f}')  # means How many digit do you want to print
```
