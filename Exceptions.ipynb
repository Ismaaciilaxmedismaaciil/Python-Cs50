# Exceptions
#### - In Python are errors detected suring excution that ditrupt the normal flow of a program.

## Common Built-In Exceptions
- `ValueError` Recieved an argument with the right but an inappropriate value. 
- `TypeError`  Applied and operation or function to an object of inappropriate type.
- `IndexError` Attempted to access a sequance index that is out of range.
- `KeyError`   Looked fir a dictionary keys that doesnot exists. 

## Handling Exceptions
#### You handle exceptions using `try`, `except`, `else` and `finally` blocks.

```python
# Example: Handling Exceptions
try:
    num = int(input('Enter a number? '))
    result = 10 / num

except ValueError:
    print('That is not valid number!')
except ZeroDivisionError:
    print('You can not divide by zero.')
else:
    print(f'Succes! The result is {result}.')

finally:
    print('This code always rund.')
```

## Raising Exceptions
#### You can internationally triger exceptions using the `raise` keyword.

```python
# Example: Raise exception
age = -22
if age < 0:
    raise ValueError('Age can not be negative.')
print(age)
```

### Using while Loop 
#### 1 - break keyword

```python
while True:
    try:
        x = int(input('Enter the x? '))
    except ValueError:
        print('X is not an integer!')
    else:
        break
print(f'x is {x}')
```

```python
# Another attempted
while True:
    try:
        X = int(input('What is x?'))
    except ValueError:
        print(f'X is not an integer!')
    else:
        break

print(f'X is {X}')
```

#### Using definition function

```python
def main():
    x = get_int()
    print(f'X is {x}')


def get_int():
    while True:
        try:
            x = int(input('What the is X? '))
        except ValueError:
            print('X is not an integer')
        
        else:
            break

    return x
main()
```

#### 2. pass keyword

```python
def main():
    x = get_int()
    print(f'X is {x}')

def get_int():
    while True:
        try:
            return int(input('What the is X? '))
        except ValueError:
            pass
main()
```

### Using function Argument

```python
def main():
    x = get_int('What is the X?')
    print(f'X is {x}')

def get_int(prompt):
    while True:
        try:
            return int(input(prompt))

        except ValueError:
            pass

main()
```
