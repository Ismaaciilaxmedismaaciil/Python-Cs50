## Dictionaries
#### Is a built-in data structures that stores data in key-value pairs.

```python
# 1. Create a diciotnary
user = {'name': 'Ismail', 
        'Age': 22}

# 2. Access data Safely (Prevent key errors if keys is missing)
age = user.get('Age') #Returns 23
print(age)

# 3. Add Or Update data
user['Eamil'] = 'Ismail@gmail.com'  # Add new pair
user['Age'] = 50                    # Update an existing key

# 4. Remove data
user.pop('Age')       # Removes 'Age' and Returns 60
```

```python
# Accessing With Iteration
students = {
    'Ismail': 'Garowe',
    'Cali': 'Garowe',
    'Jamma': 'Garowe',
    'Jamiila': 'Kismaayo'

}

# Only keys
for student in students:
    print(student)  # Returns Only key

print('=====================')

# Only Values
for student in students:
    print(students[student]) # Returns only Values

print("======================")


# Both
for student in students:
    print(student, students[student], sep=', ')  # Both keys and values
```
