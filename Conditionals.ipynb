```python
# Example 1:


x = int(input('What\'s the X? '))
y = int(input('What\'s the Y? '))

# Conditionals
if x > y:
    print('X is lessthan Y')
elif x < y:
    print('X is greaterthan Y')
elif x == y:
    print('X is equalto Y')
```

### Else keyword

```python
# Example 1:


x = int(input('What\'s the X? '))
y = int(input('What\'s the Y? '))

# Conditionals
if x > y:
    print('X is lessthan Y')
elif x < y:
    print('X is greaterthan Y')
else:
    print('X is equalto Y')
```

### Or Operator

```python
x = 2
y = 3

if x < y or x > y:
    print('X is not equal to Y')
else:
    print('X is equal to Y')
```

### Not Equal Operator

```python
x = 3
y = 5

if x != y:
    print('X is not equal to Y')
else:
    print('X is equal to Y')
```

### And Operator

```python
# Example:
score = int(input('Enter Score: '))

# Conditionals with and Operator
if score >= 90 and score <=100:
    print('Grade: A')
elif score >= 80 and score < 90:
    print('Grade: B')
elif score >= 70 and score < 80:
    print('Grade: C')
elif score >= 60 and score < 70:
    print('Grade: D')
else:
    print('Grade: F')
```
