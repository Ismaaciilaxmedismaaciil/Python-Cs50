### Else keyword

```python
# Example 1:


x = int(input('What\'s the X? '))
y = int(input('What\'s the Y? '))

# Conditionals
if x > y:
    print('X is lessthan Y')
elif x < y:
    print('X is greaterthan Y')
else:
    print('X is equalto Y')
```

### Or Operator

```python
x = 2
y = 3

if x < y or x > y:
    print('X is not equal to Y')
else:
    print('X is equal to Y')
```

### Not Equal Operator

```python
x = 3
y = 5

if x != y:
    print('X is not equal to Y')
else:
    print('X is equal to Y')
```

### And Operator

```python
# Example:
score = int(input('Enter Score: '))

# Conditionals with and Operator
if score >= 90 and score <=100:
    print('Grade: A')
elif score >= 80 and score < 90:
    print('Grade: B')
elif score >= 70 and score < 80:
    print('Grade: C')
elif score >= 60 and score < 70:
    print('Grade: D')
else:
    print('Grade: F')
```

#### Module

```python
try: 
    num = int(input("What's is x? "))

    if num %2==0:
        print(f'{num} is Even number.')

    else:
        print(f'{num} is Odd number.')

except:
    print('Error: Please Enter a numerical Value')
```

```python
# Using definition function
def main():
    x = int(input("What's X? "))

    if is_even(x):
        print('Even')

    else:
        print('Odd')

def is_even(n):
    if n % 2==0:
        return True
    
    else:
        return False
    
main()
```

```python
# Short way all in One line
def main():
    x = int(input("What's x? "))

    if is_even(x):
        print('Even')

    else:
        print('Odd')

def is_even(n):
    return True if n%2==0 else False

main()
```

```python
# Another short way
def main():
    x = int(input("What's x? "))

    if is_even(x):
        print('Even')

    else:
        print('Odd')

def is_even(n):
    return n % 2==0

main()
```

#### Match

```python
name = input("What's your name? ")

if name == 'Ismail':
    print(f'{name}, Welcome to my Party!')

elif name == 'Cali':
    print(f'{name}, Welcome to my Party!')

elif name == 'Jamma':
    print(f'{name}, Welcome to my Party!')

elif name == 'Fatima':
    print(f'{name}, Welcome to my Party!')

elif name == 'Saynab':
    print(f'{name}, Welcome to my Party!')

else:
    print(f'{name}, Your not Invite')
```

```python
# All in One line
name = input("What's your name? ")

if name == 'Ismail' or name == 'Cali' or name == 'Jamma' or name == 'Fatima' or name == 'Saynab':
    print(f'{name}, Welcome to my Party!')


else:
    print(f'{name}, Your not Invite')
```

Using `match` keyword

```python
name = input("What's your name? ")

match name:
    case 'Ismail':
        print(f'{name}, Welcome to my Party!')
    case 'Cali':
        print(f'{name}, Welcome to my Party!')
    case 'Jamma':
        print(f'{name}, Welcome to my Party!')
    case 'Fatima':
        print(f'{name}, Welcome to my Party!')    
    case 'Saynab':
        print(f'{name}, Welcome to my Party!')

    # Intas wixii kasoo hara
    case _:
        print(f'{name}, Your not Invite')
```

```python
# Short All in One Line
name = input("What's your name? ")

match name:
    case 'Ismail' | 'Cali' | 'Jamma' | 'Fatima' | 'Saynab':
        print(f'{name}, Welcome to my Party!')

    case _:
        print(f'{name}, Your not Invite')
```
